```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('titanic.csv')
```

```python
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df.groupby('Sex')['Survived'].mean()
```

```python
df.groupby('Pclass')['Survived'].mean()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])  # Select numeric columns only
correlation_matrix = numeric_df.corr()  # Compute correlation on numeric columns
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=df)

sns.histplot(df['Age'], kde=True)

sns.barplot(x='Pclass', y='Survived', data=df)

df['Embarked'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Embarked Distribution')
plt.show()
```
